```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python

URL = '/content/Bucharest Stock Exchange Data - YTD 2023.xlsx'

df_shares = pd.read_excel(URL)
print('Data downloaded and read into a dataframe!')
```

```python
df_shares.head()
```

```python
df_shares.info()
```

```python
df_shares.isnull().sum()
```

```python
df_investment = df_shares.groupby('Investment Name',axis=0).sum()
print(type(df_shares.groupby('Investment Name', axis=0)))
df_investment.head()
```

```python
df_investment['Issued Shares'].plot(kind='pie', figsize=(10, 10),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            shadow=True,       # add shadow
                            )

plt.title('Shares issued by the Investors')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```
